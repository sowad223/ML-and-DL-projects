```python
#1. Create two linear arrays a and b with 20 random elements each ranging from 1 to 100.
a = np.random.randint(1, 101, 20)
b = np.random.randint(1, 101, 20)
print(a)
print("---------------")
print(b)
```

```python
#2. Now change those linear arrays to two 5x4 matrices.
print("old shape")
print(a.shape)
print(b.shape)
print("new shape")
a = a.reshape(5, 4)
b = b.reshape(5, 4)
print(a)
print("---------------")
print(b)
```

```python
#3. Perform scalar multiplication with matrix a and b.
scalar_value=int(input("Enter scalar value: "))
a = a * scalar_value
b = b * scalar_value
print(a)
print("---------------")
print(b)
```

```python
#4. Now try to modify the matrix b and perform matrix multiplication with a and b to store it in c.

b=b.T

print("shape a:",a.shape)
print("new shape of b after trans:",b.shape)
print(b)
c=a@b
print(c)
```

```python
#5. Print only the column 2, 3 of matrix c. Print the whole matrix using slicing technique.
print(c[:,1:3])
print(c[:])
```

```python
#6. Find the largest element in the array, and also the index of that element.
max_value = c[0][0]
max_index = (0, 0)


for i in range(len(c)):
    for j in range(len(c[0])):
        if c[i][j] > max_value:
            max_value = c[i][j]
            max_index = (i, j)

print("\nLargest element in c:", max_value)
print("Index of largest element:", max_index)
```

```python
#7. Convert the matrix c into an one dimensional array keeping the original array unchanged.
print("original array:",c)
d = c.flatten() #creates a new array
print("--------------")
print("--------------")
print("converted array:",d)
```

**Use Matplotlib library to complete the following tasks:**

```python
#1. Generate a scatter plot graph for temperature_dhaka vs humidity_dhaka using (*) as the marker in color red.
import matplotlib.pyplot as plt
temperature_dhaka = [25,34,21,45,28,6,43,18,7,2]
humidity_dhaka = [28, 25,29,20, 26, 50, 19, 29, 52, 55]
plt.title("temp vs humidity")
plt.xlabel("temp")
plt.ylabel("humidity")
plt.grid()
plt.scatter(temperature_dhaka, humidity_dhaka, color="red", marker="*")
plt.show()
```

```python
#2. Now do a line plot for study_hours vs marks with figure size (12,8)
study_hours = [2,3,4,4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]
plt.figure(figsize=(12,8))
plt.title("study hours vs marks")
plt.xlabel("study hours")
plt.ylabel("marks")
plt.grid()
plt.plot(study_hours, marks)
plt.show()
```

```python
#3. Now draw a horizontal bar graph showing the scores in each of the subjects.


subjects = ['Maths', 'English', 'Science', 'Physics', 'Computer']
marks = [89, 90, 45, 78, 99]
fig, axs = plt.subplots(1, 3, figsize=(36, 12))
axs[0].barh(subjects, marks, color='green')
axs[0].set_title('Horizontal Bar Graph')
axs[0].set_xlabel('Marks')
axs[0].set_ylabel('Subjects')
#4. Draw a vertical bar graph where each subject score is visualized in different colors.
axs[1].bar(subjects, marks, color=['red', 'blue', 'green', 'orange', 'purple'])
axs[1].set_title('Vertical Bar Graph')
axs[1].set_xlabel('Subjects')
axs[1].set_ylabel('Marks')

#5. Show all these graphs in one graph using subplot( ) function.
axs[2].barh(subjects, marks, color='green')
axs[2].bar(subjects, marks, color=['red', 'blue', 'green', 'orange', 'purple'])
axs[2].set_title('Combined Graph')
axs[2].set_xlabel('Marks')
axs[2].set_ylabel('Subjects')

plt.tight_layout()
plt.show()
```
